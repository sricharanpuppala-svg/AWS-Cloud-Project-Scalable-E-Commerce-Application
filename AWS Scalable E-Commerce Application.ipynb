### Importing the libraries

```python
import pandas 
import numpy 
from datetime import datetime
```

```python
products = pd.DataFrame({
    "ProductID": [101, 102, 103, 104],
    "ProductName": ["Laptop", "Smartphone", "Headphones", "Smart Watch"],
    "Category": ["Electronics", "Electronics", "Accessories", "Wearables"],
    "Price": [1200, 850, 150, 300],
    "Stock": [15, 30, 50, 20]
})
print("PRODUCT INVENTORY")
display(products)
```

```python
orders = pd.DataFrame({
    "OrderID": [1001, 1002, 1003],
    "Customer": ["John", "Emma", "David"],
    "Product": ["Laptop", "Headphones", "Smartphone"],
    "Quantity": [1, 2, 1],
    "OrderDate": [
        "2026-05-01",
        "2026-05-03",
        "2026-05-05"
    ]
})
print("\n CUSTOMER ORDERS ")
display(orders)
```

```python
price_lookup = {
    "Laptop": 1200,
    "Headphones": 150,
    "Smartphone": 850
}

orders["TotalPrice"] = orders.apply(
    lambda row: row["Quantity"] * price_lookup[row["Product"]],
    axis=1
)

print("\n SALES REPORT ")
display(orders)
total_sales = orders["TotalPrice"].sum()
print(f"\nTotal Revenue Generated: €{total_sales}")
```

```python
low_stock = products[products["Stock"] < 20]
print("\n LOW STOCK ALERT ")
if not low_stock.empty:
    display(low_stock)
else:
    print("All products have a sufficient stock.")
```

```python
server_logs = pd.DataFrame({
    "Timestamp": [
        datetime.now(),
        datetime.now(),
        datetime.now()
    ],
    "Service": [
        "EC2 Instance",
        "RDS Database",
        "CloudFront CDN"
    ],
    "Status": [
        "Running",
        "Connected",
        "Active"
    ]
})

print("\n AWS CLOUD SERVICES STATUS ")
display(server_logs)
```
